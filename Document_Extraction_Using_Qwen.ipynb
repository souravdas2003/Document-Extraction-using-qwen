```python
# Complete I-9 Document Verification System for Google Colab
# With Feature Visualization and Robust Error Handling

# 1. Install required packages
!pip install -q pdf2image transformers torch pillow matplotlib pandas qwen-vl-utils

# 2. Import libraries
import os
import re
import pandas as pd
import numpy as np
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import io
import torch
import time
import warnings
warnings.filterwarnings("ignore")  # Suppress warnings

# Check if qwen-vl-utils is available and install if needed
try:
    from qwen_vl_utils import process_vision_info
    print("Using qwen-vl-utils for image processing")
except ImportError:
    print("qwen-vl-utils not available, using standard processing")
    process_vision_info = None

# Install poppler for PDF processing
!apt-get update -qq && apt-get install -y poppler-utils

class I9DocumentVerifier:
    def __init__(self):
        """Initialize the document verification system."""
        print("⏳ Loading Qwen2-VL model... This may take a few minutes...")
        start_time = time.time()

        # Load a smaller, more compatible model
        from transformers import AutoProcessor, AutoModelForVision2Seq

        model_name = "Qwen/Qwen2-VL-7B-Instruct"
        self.processor = AutoProcessor.from_pretrained(model_name)
        self.model = AutoModelForVision2Seq.from_pretrained(
            model_name,
            torch_dtype=torch.float16,
            device_map="auto"
        )

        end_time = time.time()
        print(f"✅ Model loaded successfully in {end_time - start_time:.2f} seconds!")

        # Define document categories for I-9 verification
        self.list_a_docs = [
            "U.S. Passport", "Permanent Resident Card", "Foreign passport with Form I-94",
            "Employment Authorization Document", "Passport from FSM or RMI"
        ]

        self.list_b_docs = [
            "Driver's license", "ID card", "School ID card", "Voter's registration card",
            "U.S. Military card", "Military dependent's ID card",
            "U.S. Coast Guard Merchant Mariner Card", "Native American tribal document",
            "Canadian driver's license"
        ]

        self.list_c_docs = [
            "Social Security Account Number card", "Birth certificate",
            "Form FS-545", "Form DS-1350", "U.S. Citizen ID Card",
            "ID Card for Resident Citizen", "DHS employment authorization document"
        ]

    def visualize_image_features(self, image):
        """Extract and visualize the features from an image."""
        print("Extracting image features...")

        try:
            # 1. Get raw features using the image processor
            encoding = self.processor.image_processor(images=image, return_tensors="pt")
            pixel_values = encoding.pixel_values

            # Print basic information about the extracted features
            print(f"Raw pixel values shape: {pixel_values.shape}")

            # 2. Try to get the visual features from the model directly
            try:
                # Access the vision encoder if available
                if hasattr(self.model, "vision_model"):
                    print("Extracting features using vision_model...")
                    vision_outputs = self.model.vision_model(pixel_values.to(self.model.device))

                    # Get the features
                    last_hidden_state = vision_outputs.last_hidden_state
                    print(f"Vision model features shape: {last_hidden_state.shape}")
                    print(f"Number of feature vectors: {last_hidden_state.shape[1]}")

                    # Print sample of the first feature vector
                    print("\nSample of first feature vector (first 10 values):")
                    print(last_hidden_state[0, 0, :10].cpu().detach().numpy())

                    # Print statistics about the features
                    feature_mean = last_hidden_state.mean().item()
                    feature_std = last_hidden_state.std().item()
                    feature_min = last_hidden_state.min().item()
                    feature_max = last_hidden_state.max().item()

                    print(f"\nFeature statistics:")
                    print(f"- Mean: {feature_mean:.4f}")
                    print(f"- Std Dev: {feature_std:.4f}")
                    print(f"- Min: {feature_min:.4f}")
                    print(f"- Max: {feature_max:.4f}")

                    return last_hidden_state.shape[1]  # Return number of feature vectors
                else:
                    print("Model doesn't have a direct vision_model attribute.")
            except Exception as e:
                print(f"Error accessing vision model features: {e}")

            # 3. Alternative method to get features
            try:
                # Try processing with the model's feature extractor
                print("\nTrying alternative feature extraction...")

                # Prepare a simple prompt
                prompt = "Describe this image."
                inputs = self.processor(text=prompt, images=image, return_tensors="pt")

                # Check if we got the expected image embeddings
                if 'pixel_values' in inputs:
                    print(f"Processed pixel values shape: {inputs['pixel_values'].shape}")

                # Check for image tokens if available
                if 'image_embeds' in inputs:
                    print(f"Image embeds shape: {inputs['image_embeds'].shape}")
                    return inputs['image_embeds'].shape[1]

                # Additional debugging for token-related issues
                if 'attention_mask' in inputs:
                    print(f"Attention mask shape: {inputs['attention_mask'].shape}")
                    print(f"Input IDs shape: {inputs['input_ids'].shape}")

                    # Calculate total tokens
                    token_count = inputs['attention_mask'].sum().item()
                    print(f"Total tokens in input: {token_count}")

                return None
            except Exception as e:
                print(f"Error in alternative feature extraction: {e}")
                return None

        except Exception as e:
            print(f"Error visualizing image features: {e}")
            return None

    def upload_document(self):
        """Allow user to upload a document image."""
        print("📄 Please upload your document image or PDF...")
        uploaded = files.upload()
        if not uploaded:
            print("No files were uploaded.")
            return None

        file_path = list(uploaded.keys())[0]
        print(f"📝 Processing {file_path}...")
        return file_path

    def convert_pdf_to_image(self, pdf_path):
        """Convert a PDF file to an image."""
        try:
            print("Converting PDF to image for processing...")
            import pdf2image

            # Use a lower DPI for better performance
            images = pdf2image.convert_from_path(pdf_path, 150)
            if not images:
                raise Exception("PDF conversion returned no images")

            print(f"Successfully converted PDF to {len(images)} images")
            return images[0]  # Return the first page
        except Exception as e:
            print(f"Error converting PDF: {e}")
            return None

    def load_document(self, file_path):
        """Load document from file path, handling both images and PDFs."""
        if not file_path:
            return None

        # Check file extension
        if file_path.lower().endswith('.pdf'):
            # Handle PDF - convert to image
            image = self.convert_pdf_to_image(file_path)
            if image:
                # Save the converted image
                image_path = "converted_document.jpg"
                image.save(image_path)
                print(f"Saved converted image to {image_path}")
                return Image.open(image_path)  # Reload to ensure proper format
            return None
        else:
            # Handle regular image
            try:
                return Image.open(file_path)
            except Exception as e:
                print(f"Error opening image: {e}")
                return None

    def display_image(self, image_or_path):
        """Display the uploaded document image."""
        if isinstance(image_or_path, str):
            image = self.load_document(image_or_path)
        else:
            image = image_or_path  # Already a PIL Image

        if image:
            plt.figure(figsize=(10, 10))
            plt.imshow(image)
            plt.title("Uploaded Document")
            plt.axis('off')
            plt.show()
        return image

    def process_with_model_safely(self, prompt, image):
        """Process text and image with Qwen-VL model using a safe approach."""
        try:
            print(f"Processing with prompt: '{prompt[:50]}...'")

            # First, examine the image features to help diagnose issues
            feature_count = self.visualize_image_features(image)
            print(f"Detected {feature_count} feature vectors in the image")

            # Method 1: Standard direct approach
            try:
                # Format the inputs directly
                inputs = self.processor(
                    text=prompt,
                    images=image,
                    return_tensors="pt"
                ).to(self.model.device)

                # Generate with modest parameters
                with torch.no_grad():
                    output = self.model.generate(
                        **inputs,
                        max_new_tokens=256,
                        do_sample=False
                    )

                # Decode the output
                result = self.processor.batch_decode(output, skip_special_tokens=True)[0]

                # Clean up the result
                if "assistant" in result.lower():
                    parts = result.lower().split("assistant")
                    result = parts[-1]
                    if ":" in result:
                        result = result.split(":", 1)[1].strip()

                print("Successfully processed with Method 1")
                return result
            except Exception as e1:
                print(f"Method 1 failed: {e1}, trying Method 2...")

                # Method 2: Using qwen-vl-utils
                if process_vision_info is not None:
                    try:
                        messages = [
                            {
                                "role": "user",
                                "content": [
                                    {"type": "image", "image": image},
                                    {"type": "text", "text": prompt}
                                ]
                            }
                        ]

                        # Apply chat template
                        text = self.processor.apply_chat_template(
                            messages, tokenize=False, add_generation_prompt=True
                        )

                        # Process vision info
                        image_inputs, video_inputs = process_vision_info(messages)

                        # Prepare inputs
                        inputs = self.processor(
                            text=[text],
                            images=image_inputs,
                            padding=True,
                            return_tensors="pt"
                        ).to(self.model.device)

                        # Generate
                        with torch.no_grad():
                            output_ids = self.model.generate(
                                **inputs,
                                max_new_tokens=256,
                                do_sample=False
                            )

                        # Decode
                        result = self.processor.batch_decode(output_ids, skip_special_tokens=True)[0]
                        print("Successfully processed with Method 2")
                        return result
                    except Exception as e2:
                        print(f"Method 2 failed: {e2}, trying Method 3...")

                # Method 3: Simplest possible approach
                try:
                    encoding = self.processor.image_processor(images=image, return_tensors="pt")
                    pixel_values = encoding.pixel_values.to(self.model.device)

                    # Process text
                    input_ids = self.processor.tokenizer(prompt, return_tensors="pt").input_ids.to(self.model.device)

                    # Generate
                    with torch.no_grad():
                        output_ids = self.model.generate(
                            input_ids=input_ids,
                            pixel_values=pixel_values,
                            max_new_tokens=256,
                            do_sample=False
                        )

                    # Decode
                    result = self.processor.tokenizer.decode(output_ids[0], skip_special_tokens=True)
                    print("Successfully processed with Method 3")
                    return result
                except Exception as e3:
                    # Method 4: Last resort - resize the image drastically
                    try:
                        print("All methods failed, trying with a resized image...")
                        # Resize to a smaller size to reduce complexity
                        resized_img = image.resize((224, 224), Image.LANCZOS)

                        # Try the standard approach again with the resized image
                        inputs = self.processor(
                            text=prompt,
                            images=resized_img,
                            return_tensors="pt"
                        ).to(self.model.device)

                        with torch.no_grad():
                            output = self.model.generate(
                                **inputs,
                                max_new_tokens=256,
                                do_sample=False
                            )

                        # Decode the output
                        result = self.processor.batch_decode(output, skip_special_tokens=True)[0]
                        print("Successfully processed with resized image")
                        return result
                    except Exception as e4:
                        # All methods failed
                        print(f"All methods failed: {e1} -> {e2} -> {e3} -> {e4}")
                        return f"Unable to process image. Error: {e4}"
        except Exception as e:
            print(f"General processing error: {e}")
            return f"Processing error: {e}"

    def identify_document(self, image_or_path):
        """Identify what type of document was uploaded."""
        prompt = """
        Please analyze this image and identify what type of document it is.
        Focus specifically on I-9 acceptable documents, which include:

        List A Documents (establish both identity and employment authorization):
        - U.S. Passport or Passport Card
        - Permanent Resident Card (Form I-551)
        - Foreign passport with Form I-94
        - Employment Authorization Document with photo (Form I-766)
        - Foreign passport with Form I-94 or Form I-94A
        - Passport from FSM or RMI

        List B Documents (establish identity only):
        - Driver's license or ID card
        - School ID card with photograph
        - Voter's registration card
        - U.S. Military card or draft record
        - Military dependent's ID card
        - U.S. Coast Guard Merchant Mariner Card
        - Native American tribal document
        - Canadian driver's license

        List C Documents (establish employment authorization only):
        - Social Security Account Number card
        - Birth certificate
        - Form FS-545
        - Form DS-1350
        - U.S. Citizen ID Card (Form I-197)
        - ID Card for Resident Citizen (Form I-179)
        - DHS employment authorization document

        Respond with the document category (List A, B, or C) and specific document type.
        """

        # Get the image
        if isinstance(image_or_path, str):
            image = self.load_document(image_or_path)
        else:
            image = image_or_path  # Already a PIL Image

        if not image:
            return "Failed to load document image"

        # Process with model
        document_info = self.process_with_model_safely(prompt, image)

        print("📋 Document Identification:\n")
        print(document_info)
        return document_info

    def extract_document_info(self, image_or_path, doc_type):
        """Extract detailed information from the document."""
        prompt = f"""
        This image shows a {doc_type}. Please extract all relevant information from this document
        and format it in a structured way.

        For each piece of information, provide both the field name and its value.
        Common fields include:

        - Document Type
        - First Name / Given Names
        - Last Name / Surname
        - Date of Birth
        - Date of Issue
        - Date of Expiration
        - Document Number
        - Issuing Authority
        - Country/State
        - Sex/Gender

        For a passport, also include:
        - Nationality
        - Place of Birth
        - MRZ (Machine Readable Zone)

        For a driver's license, also include:
        - Address
        - Class/Type
        - Restrictions (if any)

        Format your response as a structured list with fields and values clearly labeled, like:
        **Document Type**: [value]
        **Country**: [value]
        **Last Name**: [value]
        etc.
        """

        # Get the image
        if isinstance(image_or_path, str):
            image = self.load_document(image_or_path)
        else:
            image = image_or_path  # Already a PIL Image

        if not image:
            return "Failed to load document image"

        # Process with model
        extracted_info = self.process_with_model_safely(prompt, image)

        print("📃 Extracted Information:\n")
        print(extracted_info)
        return extracted_info

    def extract_expiration(self, text):
        """Extract expiration date from text if present."""
        expiration_pattern = r'(?:expir|expiration|expires|exp date|exp).*?(\d{1,2}[\/\-\s\.]\d{1,2}[\/\-\s\.]\d{2,4}|\d{1,2}[\/\-\s\.][A-Za-z]{3,}[\/\-\s\.]\d{2,4})'
        match = re.search(expiration_pattern, text.lower())
        if match:
            return match.group(1)
        return "Not found"

    def check_document_validity(self, extracted_info):
        """Check if document appears to be valid based on expiration."""
        expiration = self.extract_expiration(extracted_info)
        if expiration == "Not found":
            return "Unable to determine expiration date"

        # This is simplified logic - would need more complex date parsing in production
        if "expired" in extracted_info.lower():
            return "⚠️ Document appears to be EXPIRED"

        return "✅ Document appears to be VALID (based on available information)"

    def convert_to_dataframe(self, extracted_text):
        """Convert the extracted text information to a pandas DataFrame."""
        lines = extracted_text.split('\n')
        data = {}

        for line in lines:
            if '**' in line and ':' in line:
                # Extract field and value
                parts = line.split(':')
                if len(parts) >= 2:
                    field = parts[0].replace('*', '').strip()
                    value = ':'.join(parts[1:]).strip()
                    data[field] = [value]

        # Create DataFrame
        if not data:  # If no structured data was found
            data = {"Raw Text": [extracted_text]}

        # Create DataFrame
        df = pd.DataFrame(data)
        return df

    def check_i9_eligibility(self, doc_info):
        """Check I-9 eligibility based on document type."""
        doc_info_lower = doc_info.lower()

        # Check for List A documents (establish both identity and work authorization)
        if any(doc.lower() in doc_info_lower for doc in self.list_a_docs):
            return "✅ LIST A DOCUMENT: Establishes both identity and employment authorization"

        # Check for List B documents (establish identity only)
        elif any(doc.lower() in doc_info_lower for doc in self.list_b_docs):
            return "⚠️ LIST B DOCUMENT: Establishes identity only - need a List C document as well"

        # Check for List C documents (establish work authorization only)
        elif any(doc.lower() in doc_info_lower for doc in self.list_c_docs):
            return "⚠️ LIST C DOCUMENT: Establishes employment authorization only - need a List B document as well"

        else:
            return "❌ UNKNOWN DOCUMENT TYPE: Cannot determine I-9 eligibility"

    def verify_document(self, file_path):
        """Main function to verify a document."""
        # Load and display the document
        image = self.load_document(file_path)
        if not image:
            print("❌ Failed to load document. Please check the file format.")
            return None

        # Display the image
        self.display_image(image)

        # Identify document type
        doc_info = self.identify_document(image)

        # Extract document type from the model's response
        doc_type = "Unknown"
        for line in doc_info.split('\n'):
            if any(doc_name.lower() in line.lower() for doc_list in [self.list_a_docs, self.list_b_docs, self.list_c_docs]
                   for doc_name in doc_list):
                doc_type = line.strip()
                break

        # Extract detailed information
        extracted_info = self.extract_document_info(image, doc_type)

        # Convert to DataFrame
        df = self.convert_to_dataframe(extracted_info)

        # Check document validity
        validity = self.check_document_validity(extracted_info)

        # Check I-9 eligibility
        eligibility = self.check_i9_eligibility(doc_info)

        return {
            "document_type": doc_type,
            "extracted_info": extracted_info,
            "dataframe": df,
            "validity": validity,
            "eligibility": eligibility
        }

    def export_results(self, result, format="csv"):
        """Export the results to the specified format."""
        if format == "csv":
            result["dataframe"].to_csv('document_verification_result.csv', index=False)
            print("\n💾 Results saved to document_verification_result.csv")
            files.download('document_verification_result.csv')

        elif format == "excel":
            result["dataframe"].to_excel('document_verification_result.xlsx', index=False)
            print("\n💾 Results saved to document_verification_result.xlsx")
            files.download('document_verification_result.xlsx')

# Run the document verification system
def main():
    print("=" * 80)
    print("🔍 I-9 DOCUMENT VERIFICATION SYSTEM")
    print("   Using Qwen2-VL model with Transformers")
    print("=" * 80)

    # Initialize the verifier
    verifier = I9DocumentVerifier()

    # Upload and process document
    file_path = verifier.upload_document()
    if not file_path:
        return

    print("\n⏳ Processing document...")
    result = verifier.verify_document(file_path)

    if not result:
        print("\n❌ Document processing failed.")
        return

    # Display summary
    print("\n" + "=" * 80)
    print("📊 DOCUMENT VERIFICATION SUMMARY")
    print("=" * 80)

    print(f"📄 Document Type: {result['document_type']}")
    print(f"🔐 Validity: {result['validity']}")
    print(f"📋 I-9 Eligibility: {result['eligibility']}")

    # Display data in a table
    print("\n📋 Extracted Information (Table Format):")
    display(result["dataframe"])

    # Export options
    print("\n💾 Export options:")
    export_format = input("Export to CSV or Excel? (csv/excel/none): ").lower()
    if export_format in ["csv", "excel"]:
        verifier.export_results(result, format=export_format)

    print("\n✅ Document verification complete!")

# Test just feature visualization independently
def test_feature_visualization():
    print("=" * 80)
    print("🔍 IMAGE FEATURE VISUALIZATION TEST")
    print("=" * 80)

    # Initialize the verifier
    verifier = I9DocumentVerifier()

    # Upload and process document
    print("📄 Please upload an image or PDF to visualize features...")
    uploaded = files.upload()
    if not uploaded:
        print("No files were uploaded.")
        return

    file_path = list(uploaded.keys())[0]
    print(f"📝 Processing {file_path}...")

    # Load the document
    image = verifier.load_document(file_path)
    if not image:
        print("❌ Failed to load document. Please check the file format.")
        return

    # Display the image
    verifier.display_image(image)

    # Visualize the features
    feature_count = verifier.visualize_image_features(image)

    if feature_count is not None:
        print(f"\n✅ Successfully extracted {feature_count} feature vectors from the image.")
    else:
        print("\n❌ Could not extract features from the image.")

# Choose which mode to run
run_mode = "full"  # Change to "features" to just visualize features

if __name__ == "__main__":
    if run_mode == "features":
        test_feature_visualization()
    else:
        main()
```
